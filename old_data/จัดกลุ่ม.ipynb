```python
import pandas as pd
df = pd.read_excel("sec3.xlsx")
print(len(df))
df_sorted = df.sort_values(by='เกรดเฉลี่ย').reset_index(drop=True)
# Create empty lists for the groups (10 groups)
groups = [[] for _ in range(10)]

# Split the sorted DataFrame into two halves
midpoint = len(df_sorted) // 2
lower_half = df_sorted.iloc[:midpoint]  # Bottom half (lower GPAs)
upper_half = df_sorted.iloc[midpoint:]  # Top half (higher GPAs)

# Define group sizes: 3 groups of 6 students, and 7 groups of 5 students
group_sizes = [6] * 3 + [5] * 7  # 3 groups with 6 students, 7 groups with 5 students

# Shuffle group sizes so they don't cluster the larger groups together
import random
random.shuffle(group_sizes)

# Ensure we are distributing students alternately between top and bottom halves
group_assignments = []
group_idx = 0

# Distribute students from both halves evenly across the groups
for i in range(midpoint):
    # Add one student from the upper half to the current group
    if len(groups[group_idx]) < group_sizes[group_idx]:
        groups[group_idx].append((upper_half.iloc[i]['รหัส'], upper_half.iloc[i]['เกรดเฉลี่ย']))
        group_assignments.append((upper_half.iloc[i]['รหัส'], group_idx + 1))  # Assign to group starting from 11

    # Add one student from the lower half to the current group
    if len(groups[group_idx]) < group_sizes[group_idx]:
        groups[group_idx].append((lower_half.iloc[i]['รหัส'], lower_half.iloc[i]['เกรดเฉลี่ย']))
        group_assignments.append((lower_half.iloc[i]['รหัส'], group_idx + 1))  # Assign to group starting from 11
    
    # If the current group is full, move to the next group
    if len(groups[group_idx]) == group_sizes[group_idx]:
        group_idx += 1

# If there are any remaining students in the upper half (if the number of students is odd)
if len(upper_half) > len(lower_half):
    remaining_student = upper_half.iloc[-1]
    if len(groups[group_idx]) < group_sizes[group_idx]:
        groups[group_idx].append((remaining_student['รหัส'], remaining_student['เกรดเฉลี่ย']))
        group_assignments.append((remaining_student['รหัส'], group_idx + 1))  # Assign the last student to a group

# Display the groups with average GPA
for i, group in enumerate(groups, 1):  # Start enumerating from Group 11
    if len(group) > 0:  # Ensure no empty groups
        total_gpa = sum(gpa for _, gpa in group)
        avg_gpa = total_gpa / len(group)
        
        print(f"Group {i}:")
        for student_id, gpa in group:
            print(f"  Student ID: {student_id}, GPA: {gpa}")
        print(f"  Average GPA for Group {i}: {avg_gpa:.2f}")
        print()  # Blank line between groups

group_df = pd.DataFrame(group_assignments, columns=['รหัส', 'กลุ่ม'])
# Merge the group assignments back to the original sorted DataFrame
df_final = pd.merge(df_sorted, group_df, on='รหัส')
df_final = df_final.drop(columns=['เกรดเฉลี่ย'])
# Save to an Excel file
df_final.to_excel('sec3group.xlsx', index=False)
```

```python
import pandas as pd

df = pd.read_excel("sec1.xlsx")
print(len(df))

# Step 2: Sort by GPA
df_sorted = df.sort_values(by='เกรดเฉลี่ย', ascending=False).reset_index(drop=True)

# Steps 3 & 4: Initialize groups and assign students
num_groups = 10
groups = [[] for _ in range(num_groups)]
group_gpa_sums = [0.0 for _ in range(num_groups)]

for index, row in df_sorted.iterrows():
    # Find the group with the minimum total GPA sum
    min_gpa_group_index = group_gpa_sums.index(min(group_gpa_sums))
    
    # Assign the student to this group (groups are numbered from 1 to num_groups)
    df_sorted.at[index, 'Group'] = min_gpa_group_index + 1  # +1 to start group numbering from 1
    
    # Update the group's total GPA sum
    group_gpa_sums[min_gpa_group_index] += row['เกรดเฉลี่ย']

# Step 5: Convert groups to DataFrames
group_dfs = [pd.DataFrame(group) for group in groups]
df_with_groups = df.merge(df_sorted[['รหัส', 'Group']], on='รหัส')
# Step 6: Verify group sizes and averages
for group_num in range(1, num_groups + 1):
    group_df = df_with_groups[df_with_groups['Group'] == group_num]
    group_size = len(group_df)
    average_gpa = group_df['เกรดเฉลี่ย'].mean()
    print(f"Group {group_num}: Size = {group_size}, Average GPA = {average_gpa:.2f}")
# Save to an Excel file
df_with_groups = df_with_groups.drop(columns=['เกรดเฉลี่ย'])
df_with_groups.to_excel('sec1group.xlsx', index=False)
```
